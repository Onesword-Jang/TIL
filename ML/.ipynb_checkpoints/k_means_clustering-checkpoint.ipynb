# k-means-clustering
- 데이터를 그림으로 나타냈을 때 가까운 군집으로 나타내는 것을 말한다.

## 1. 알고리즘 단계
1. 초기화: k개를 임의로 설정해 중심을 잡는다.
2. 할당 단계: 데이터들을 보고 가장 가까운 중심데이터로 할당한다.
3. 업데이트 단계: 군집의 중심을 해당 군집에 속한 데이터 포인트들의 평균으로 업데이트
4. 반복: 할당과 업데이트를 변하지 않을 때까지 반복한다.

## 2. 최적의 k를 선택하는 방법: 엘보우 방법
- 군집 응집도를 계산한다.
- 이를 그래프로 나타내고 응집도가 급격히 감소하는 지점을 찾고 k로 지정한다.

## 3. k-maens 알고리즘의 단점
- 중심을 찾아야한다.
- 모든 데이터의 거리 찾기 때문에 오래걸린다.

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('Mall_Customers.csv')

data = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

```

```python
# 엘보우 방법

#inertia = 탄성, 관성(번역)
inertia = []
k = range(1,11)
for k in k:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_scaled)
    inertia.append(kmeans.inertia_)
    
    
# 그래프 그리기
plt.figure(figsize=(10,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()

kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(data_scaled)

data['Cluster'] = kmeans.labels_
```

```python
# KMeans 모델 생성 및 군집 추가
kmeans = KMeans(n_clusters=5, random_state=42)  # 군집 개수(n_clusters)는 조정 가능
# fit_predict 메서드는 데이터를 모델에 적합(fit)시키고, 각 데이터 포인트가 속한 군집을 예측해 반환
# 반환된 군집 번호는 Cluster라는 새 열로 추가되어 데이터프레임에 저장
data['Cluster'] = kmeans.fit_predict(data_scaled)  # 클러스터 결과 추가

# 2차원으로 군집 시각화 (연령 vs 소득)
plt.figure(figsize=(10, 8))
# scatterplot은 산점도를 생성하는 함수
# hue=data['Cluster']는 Cluster 열에 따라 색상을 구분하여 각 군집을 시각적으로 구분
# palette='viridis'는 색상 팔레트를 지정하여 데이터 포인트의 색상을 나타낸다.
sns.scatterplot(x=data['Age'], y=data['Annual Income (k$)'], hue=data['Cluster'], palette='viridis')
plt.title('Clusters of customers (Age vs Annual Income)')
plt.show()

# 2차원으로 군집 시각화 (소득 vs 지출 점수)
plt.figure(figsize=(10, 8))
sns.scatterplot(x=data['Annual Income (k$)'], y=data['Spending Score (1-100)'], hue=data['Cluster'], palette='viridis')
plt.title('Clusters of customers (Annual Income vs Spending Score)')
plt.show()
```
