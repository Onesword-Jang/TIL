# 아스키(ASCII) American Standard Code for Information Interchange


## ASCII 란?
- 문자와 숫자를 매핑하는 표준 코드 체계입니다. 
- 컴퓨터가 문자를 숫자로 저장하고 처리할 수 있도록 설계된 초기 문자 인코딩 시스템입니다. 


## 아스키 코드의 범위
- 아스키 코드는 총 128개의 문자로 구성
- 각 문자는 7비트로 표현된다(0127).
- 확장 아스키(Extended ASCII)는 8비트를 사용하며 256개(0255)까지 표현할 수 있다.

### 1. 제어 문자 (0~31 및 127):
- 출력되지 않는 특별한 문자로, 데이터를 제어하기 위해 사용한다.
 - 예: \n (줄 바꿈, 10), \t (탭, 9), \r (캐리지 리턴, 13), \0 (널 문자, 0).
 - 캐리지 리턴: 파이썬에서는 덮어씌우는 용도로 주로 쓰임, 본래는 줄 바꿈 없이 커서를 줄의 처음으로 이동시킨다.
 - 캐리지 리턴 활용: 덮어쓰기, 프로그래스 바
 ```python
# 덮어 쓰기
print("Hello World!\rGoodbye")  # 출력: "Goodbye World!"

# 프로그래스 바
import time

for i in range(101):
    print(f"Progress: {i}%", end='\r')
    time.sleep(0.1)
```


### 2. 출력 가능한 문자 (32~126):
- 공백, 숫자, 대문자, 소문자, 기호 등이 포함된다.
- 예:
 - 공백: 32
 - 숫자: '0''9' (4857)
 - 대문자: 'A''Z' (6590)
 - 소문자: 'a''z' (97122)
 - 기호: !, @, #, ...

## 아스키 코드 확인 방법
### 파이썬에서 아스키 코드 사용
- ord 함수: 문자를 아스키 코드 값으로 변환.

```python
코드 복사
print(ord('A'))  # 출력: 65
print(ord('a'))  # 출력: 97
print(ord('0'))  # 출력: 48
```
- chr 함수: 아스키 코드 값을 문자로 변환.

```python
코드 복사
print(chr(65))  # 출력: A
print(chr(97))  # 출력: a
print(chr(48))  # 출력: 0
```

## 예제: 아스키 코드 테이블 출력
- 다음은 32~126 범위의 출력 가능한 ASCII 문자를 테이블로 출력하는 코드입니다.

```python
for i in range(32, 127):
    print(f"{i}: {chr(i)}")
```

- 결과 (일부):

```makefile
32:  
33: !
34: "
35: #
...
65: A
66: B
...
97: a
98: b
...
126: ~
```

## 주요 활용
1. 데이터 전송: 초기 네트워크 통신에서 주로 사용.
2. 문자 변환: 대문자 ↔ 소문자 변환: ord와 chr을 사용해 간단히 계산.

```python
char = 'A'
lower = chr(ord(char) + 32)  # 'a'
print(lower)
```
3. 암호화/복호화:
 - 간단한 암호화 알고리즘에서 문자를 숫자로 변환할 때 사용.
 - 아스키는 유니코드(Unicode)가 널리 사용되기 이전의 문자 표준이며, 여전히 많은 응용 프로그램에서 사용됩니다.
