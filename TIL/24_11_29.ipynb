## 마음가짐 및 회고
오늘 스트림릿 구현을 무조건 구현할 생각입니다.

백엔드 팀원과 변경 될 내용의 공유가 필수적이라 생각하지만 아침에 내용을 공유했는데 본인이 구현한 프론트부분을 그대로 사용하라는 말은 충격이였습니다.

그건 팀 프로젝트가 아니라 본인의 개인 프로젝트를 예쁘게 꾸미기나 하라라는 말로 들렸고 우리가 처음 목표는 완성도를 높이자고 말했지만 과제의 의도가 학습이 목적이고 팀 프로젝트의 흐름을 파악하고 체험하는게 원초적인 목적이라 생각했습니다.

백엔드를 맡은 분이 구현한 프론트부분이 완벽하다고해도 우리는 이게 뭔지도 모르고 우리가 맡은 부분에서 우리의 지분 자체가 없는데 우리가 가져갈 수 있는게 없다고 생각해 분명하게 의사를 전달해서 조율을 하였습니다.

이렇게 조율했지만 우리가 생각하는 구현할 기능을 사용하기 위해선 기존의 백엔드 팀원분의 스트림릿 코드를 바꿀 수 밖에 없는 상황에서 이렇게해야한다. 공유를하니 본인 코드가 바뀌는게 원하지않는다고 말을했는데 이것도 충격이였습니다.

우리가 왜 조율을했는지 말만하고 본인이 원하는대로만 하고싶어하시는거같아서 답답한 마음뿐입니다.

우선 내가 학습할 부분은 학습하고 구현하고싶은 부분을 구현할 때 코드가 바뀌면 알리고 상대방이 너무 싫다고하면 확실하게 말을해야할듯 합니다.

## streamlit 구현
오늘 구현하며 바뀌게된 내용 및 알게된 내용을 공유하겟습니다.

- st.set_page_config()는 스크립트 최상단에 위치시켜야 합니다.

- 구현 필요: 카테고리 선택에 카테고리를 선택해주세요 추가
 - 구현 완료(NONE파일을 만들어서 경로 추가/아무 작동하지 않게 만들기 위해서)

- 프롬프트 저장을 txt, yaml파일 중 yaml을 사용한 이유
 - 데이터 관리: YAML은 계층적 데이터 구조를 표현할 수 잇어 복잡한 설정을 할 때 유용
 - 가독성
 - 필요: pip install pyyaml

- 튜터님에게 상읳서 프롬프트를 외부에서 불러오는게 아니라 코드에 포함시켜서 사용하는 방향으로 결정

- 로그인 
 - 바로 로그인 창을 띄우는 방식 (옵션 1):
```
장점:
앱을 실행하자마자 로그인/회원가입 화면이 나타나므로 보안 관점에서 좋습니다.
사용자가 명확하게 로그인 프로세스를 이해할 수 있습니다.
단점:
로그인 절차가 필수적이라 처음 접속 시 진입 장벽이 생길 수 있습니다.
사용자가 로그인 없이 탐색하거나 면접을 체험할 기회가 없습니다.

로그인 버튼을 만들어서 면접 시작 전에 로그인 (옵션 2):
장점:
로그인 없이 앱의 주요 화면을 탐색할 수 있어 사용자 경험이 더 유연합니다.
로그인 절차가 면접 기능에 진입하기 직전에만 필요하므로 부담이 줄어듭니다.
단점:
로그인 없이 접근 가능한 화면에서 보안 취약점이 생길 가능성이 있습니다.

```
- 로그인 정보를 다른 페이지에서 가져오고 확인하기.
```
main_page.py에서 로그인 정보를 json에 저장한 상태인데 
페이지 스위치로 Provide_Question.py에서 이 유저 아이디를 어떻게 가져오고 같은 사용자인지 어떻게 확인하나요?
```

제공 받은 코드와 차이가 생겼지만 구현을 완료하였고 내요을 공유하였습니다.
