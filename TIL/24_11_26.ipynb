현재 프로젝트의 흐름이 조금 빠그라졌다? 라고 생각합니다.

의견이 나오고 회의 때마다 좋다고 생각하는 쪽으로 의견수렴하는 형태로 진행하였는데 현재 가닥이 조금씩 잡히면서 이렇게할껄 이런 생각이 많이 든다고 생각합니다.
