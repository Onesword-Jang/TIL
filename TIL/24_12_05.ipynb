#### ë¬¸ì œ 2
ë¦¬ìŠ¤íŠ¸ numbers = [1, 2, 3, 4, 5, 6]ì—ì„œ ì§ìˆ˜ë§Œì„ í¬í•¨í•˜ëŠ” ìƒˆë¡œìš´ ë¦¬ìŠ¤íŠ¸ë¥¼ ì¶œë ¥í•˜ì„¸ìš”
```python
numbers = [1,2,3,4,5,6]
number = []
for x in numbers:
    if x % 2 == 0:
        number.append(x)
print(number)

numbers = [1,2,3,4,5,6]
number = [x for x in numbers if x%2==0]
print(number)
```

#### ë¬¸ì œ 3
2ì°¨ì› ë¦¬ìŠ¤íŠ¸ matrix = [[1, 2], [3, 4], [5, 6]]ì—ì„œ ê° ìš”ì†Œë¥¼ ëª¨ë‘ ë”í•œ í•©ê³„ë¥¼ ê³„ì‚°í•˜ì„¸ìš”.
```python
matrix = [[1, 2], [3, 4], [5, 6]]
result = 0
for x in matrix:
    for num in x:
        result+=num
print(result)

matrix = [[1, 2], [3, 4], [5, 6]]
result = 0
for x in matrix:
    result += sum(x)
print(result)

matrix = [[1, 2], [3, 4], [5, 6]]
result = sum([item for row in matrix for item in row])
print(result)
```
