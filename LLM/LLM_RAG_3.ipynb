# 3-1 Shot 계열의 프롬프팅 기법

## Shot이란?
- 인공지능에게 전달한 예제를 의미
- **예제를 하나 주면**: **One-Shot**
- **예제를 안 주면**: **Zero-Shot**
- **여러 개 주면**: **Few-Shot**

## 1. Zero-Shot 프롬프팅
- 예제를 주지 않고 지침만 전달하는 방식

> System: 너는 전달된 구구단에 대한 농담을 해야해.

> User: 6 * 3

> Assistant: 왜 6은 3을 곱하면 항상 놀라는 표정을 짓는 걸까요?
왜냐하면 답이 18이라고 귓속말로 말해주니까요!

## 2. One-Shot 프롬프팅
- 예제를 하나 전달해주는 방식

> System: 너는 구구단 농담을 하는 인공지능이야. 이건 구구단 농담의 예시야.
"""
6 * 3: 육삼빌딩
"""

> User: 2 * 4

> Assistant: 2 * 4: 이사오세요

## 3. Few-Shot 프롬프팅
- 여러 개의 예시를 줘서, 모델이 더 잘 학습하게 한다.
```python
System: 너는 예를 참고해서 전달된 구구단에 대한 농담을 해야해. 과정은 말하지 말고 농담만 말해.
"""
입력 : 6 * 4
출력: 육사생도
"""
"""
입력: 2 * 8
출력: 이팔 청춘
"""
"""
입력: 5 * 2
출력: 오이팩
"""
```

> User: 2 * 4

> Assistant: 이사갈 시간

## 4. 추가 팁
- Few-Shot 프롬프팅을 할 때,입력과 출력만 제공하는 것보다 과정도 함께 제공하면 좋다.

```pyton
System: 너는 예를 참고해서 전달된 구구단에 대한 농담을 해야해. 과정은 말하지 말고 농담만 말해.
"""
입력 : 6 * 4
출력: 육사생도
과정: 6 * 4는 한국어로 '육곱하기사'야. 여기에서 '곱하기'를 빼면 '육사'가 되지? '육사'로 시작하는 단어를 찾은 거야.
"""
"""
입력: 2 * 8
출력: 이팔 청춘
과정: 2 * 8은 한국어로 '이곱하기팔'이야. 여기서 '곱하기'를 빼면 '이팔'이 되지? 이팔로 시작하는 단어를 찾은 거야.
"""
"""
입력: 5 * 2
출력: 오이팩
과정: 5 * 2는 '오곱하기이'야. 여기서 '곱하기'를 빼면 '오이'가 돼. '오이'로 시작하는 단어를 찾은 거야.
"""
User: 2 * 4
Assistant: 이사오세요
```

# 3-2 Act As 류의 프롬프팅 기법

## 1. 페르소나(Persona)란?
- LLM에게 성격, 관점, 또는 역할을 부여하는 것

### 1) Act As 프롬프팅 기법
- 역할 부여 프롬프트를 의미
- 특정 직업이나 여할 부여
- 직업이나 성격에 맞는 답변을 하게 만든다.

### 2) 페르소나 설정의 구체화
- 구체적으로 지침을 주면 역할의 특성을 더 잘 반영한다.

### 3) 페르소나 창조하기
- 가상의 인물에게 성격과 직업을 설정하고 그 인물의 관점에서 답변하게 만들 수 있다.

### 4) 도구의 페르소나 부여하기
- 프로그램이나 도구에도 페르소나를 부여할 수 있다.
- ex) Python 인터프리터 처럼 대답해줘

### 5) 논리적인 추론 강화하기
- 논리적인 추론 능력이 상대적으로 부족할 수 있다.
- 구체적인 지침을 줘서 예방 가능

# 3-3 논리적인 추론 강화

## 1. 논리적 추론의 한계
- LLM은 문맥을 기반으로 확률적인 답변을 제공한다.
- 수학 문제나 논리적 퍼즐 같은 문제에서 잘못된 답변이 올 수 있다.

## 2. CoT(Chain of Thought, 생각의 사슬) 기법
- 단계별로 추론 과정을 설명하도록 모델에게유도하는 기법
- 중간 추론 단계를 명시하면, 모델이 복잡한 문제를 더 정확히 풀 수 있게 된다.

### 1) Cot 적용 예시
```python
System: 너는 문제를 풀 때 중간 과정을 설명해야 해.
User: 자동차가 3대 있어, 오토바이는 12대 있어. 이중 바퀴가 모두 터진 장비는 총 3대야. 그런데 터진 바퀴가 총 12개면 자동차와 오토바이는 각각 몇 대 고장났을까?
Assistant: 문제를 해결하기 위해서는 각 장비의 바퀴 수와 터진 바퀴의 총 수를 고려해야 해.
자동차는 4개의 바퀴가 있고, 오토바이는 2개의 바퀴가 있어.
이 문제에서는 총 3대의 장비가 바퀴가 모두 터져 있고, 터진 바퀴는 12개야.

먼저, 자동차가 x대, 오토바이가 y대가 고장났다고 가정하자.
두 가지 조건을 만족해야 해:
1. x + y = 3 (고장난 장비는 총 3대)
2. 4x + 2y = 12 (총 터진 바퀴 수는 12개)

첫 번째 식에서 y = 3 - x라고 할 수 있어.
이 값을 두 번째 식에 대입하면:
4x + 2(3 - x) = 12
4x + 6 - 2x = 12
2x = 6
x = 3

따라서 자동차 3대가 모두 고장났고, 오토바이는 고장나지 않았어.

```
### 2) 직접 중간 과정 넣기

```python
System: 너는 문제를 풀 때 중간 과정을 설명해. 중간 단계는 문제 분석, 계산, 검증의 과정을 거쳐야 해.
```

### 3) 복잡한 명령어 처리(수학 문제, 삼행시 등)
```python
System: 너는 삼행시를 만들 때, 각 줄의 의미를 단계별로 설명해야 해.
User: 나주평야로 삼행시 지어줘.
Assistant:
나: '나'는 나주평야의 '나'를 따서 시작합니다. 나주평야는 넓고 비옥한 땅을 의미하죠.
주: '주'는 주말에 나주평야를 방문한다는 뜻입니다. 자연과 함께하는 주말을 상상해볼 수 있겠네요.
평야: 평야는 넓은 들판을 상징하며, 나주평야의 풍요로움을 강조합니다. 넓고 끝없는 평화로운 공간을 말하는 것이죠.
```
