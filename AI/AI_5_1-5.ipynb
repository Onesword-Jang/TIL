````python
# 5-3 ìŒì„± ìƒì„± ElevenLabs AI ì‹¤ìŠµ

# python
import os
import requests
from pydub import AudioSegment
from pydub.playback import play
import io

# .env íŒŒì¼ì—ì„œ API í‚¤ì™€ URLì„ ê°€ì ¸ì˜¤ëŠ” ë°©ë²•
load_dotenv()

api_key = os.getenv("API_KEY")
url = os.getenv("APL_URL")

# ì„¤ì • ê°€ëŠ¥ ë³€ìˆ˜
output_filename = "output_audio.mp3"

headers = {
    "xi-api-key": api_key,
    "Content-Type": "application/json"
}

# ì½”ë“œë¡œ APIí‚¤ì™€ URLê°€ì ¸ì˜¤ê¸°
output_filename = "output_audio.mp3"

url = "ëª¨ë¸ URL"
headers = {
    "xi-api-key": "API - KEY",
    "Content-Type": "application/json"
}

# ì‚¬ìš©ìì—ê²Œ ì…ë ¥ ë°›ê¸°
text = input("í…ìŠ¤íŠ¸ë¥¼ ì…ë ¥í•˜ì„¸ìš”: ")

# ìŒì„± ë°ì´í„° ìš”ì²­
data = {
    "text": text,
    "model_id": "eleven_multilingual_v2",
    "voice_settings": {
        # ì–¼ë§ˆë‚˜ ë³€í™”ì—†ì´ ì•„ë‚˜ìš´ì„œ ì²˜ëŸ¼ ì½ì€ ê²ƒì¸ì§€(0.3 ì´í•˜ëŠ” ì¶”ì²œí•˜ì§€ ì•ŠìŒ)
        "stability": 0.3,
        # ëª©ì†Œë¦¬ì˜ ìœ ì‚¬ë„ë¥¼ ì„¤ì •
        "similarity_boost": 1,
        # ì„¤í•˜ëŠ” ì‚¬ëŒì˜ ì–µì–‘ì„ ì–¼ë§ˆë‚˜ ë”°ë¼ ê°ˆì§€(ìŠ¤íƒ€ì¼ì„ ë‚®ì¶”ì–´ì•¼ì§€ ë‹¤ë¥¸ ë‚˜ë¼ì–¸ì–´ë¥¼ ì ìš©í–ˆì„ ë•Œ ì¢‹ë‹¤.)
        "style": 1,
        "use_speaker_boost": True
    }
}

# ë¦¬í€˜ìŠ¤íŠ¸ ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ í†µí•´ hostì— í•´ë‹¹í•˜ëŠ” í˜¸ì¶œì„ í•œë‹¤.
# dataì™€ headersë¥¼ ì „ë‹¬í•œë‹¤.
# stream=TrueëŠ” ì‘ë‹µ ë°ì´í„°ë¥¼ ìŠ¤íŠ¸ë¦¬ë° ë°©ì‹ìœ¼ë¡œ ë°›ëŠ”ì§€ ì„¤ì •
response = requests.post(url, json=data, headers=headers, stream=True)

# ì‘ë‹µì½”ë“œë¥¼ ì„¤ì •
# 200ì€ HTTP ìƒíƒœì½”ë“¤ë¥´ ë‚˜íƒ€ë‚´ëŠ”ë° APIìœ í†µì´ ì„±ê³µí–ˆì„ ë•Œ 200ì´ ë‚˜íƒ€ë‚œë‹¤.
if response.status_code == 200:
    # b"" ì˜¤ë””ì˜¤ ë¬¸ìì—´ì´ ë°”ì´íŠ¸ ë‹¨ìœ„ì—¬ì„œ ë°”ì´íŠ¸ ë‹¨ìœ„ë¥¼ ì´ˆê¸°í™”í•´ì¤€ë‹¤.
    audio_content = b""
    # ì²­í¬ë‹¨ìœ„ë¡œ ì½ì–´ì™€ ì‘ë‹µë°ì´í„° ì²˜ë¦¬
    for chunk in response.iter_content(chunk_size=1024):
        if chunk:
            audio_content += chunk

    # AudioSegmentëŠ” pydubì˜ íŒ¨í‚¤ì§€
    # ë°”ì´íŠ¸ ë°ì´í„°ë¥¼ io.BytesIOë¥¼ ì‚¬ìš©í•´ì„œ ë©”ëª¨ë¦¬ íŒŒì¼ ê°ì²´ë¡œ ë³€í™˜í•œ ë’¤ 
    # from_mp3ë¥¼ ì‚¬ìš©í•´ì„œ mo3íŒŒì¼ë¡œ ë°”ê¾¼ë‹¤.
    segment = AudioSegment.from_mp3(io.BytesIO(audio_content))
    # ë³€í™˜ëœ ì˜¤ë””ì˜¤ë¥¼ ì§€ì •í•œ íŒŒì¼ ì´ë¦„, í˜•ì‹ìœ¼ë¡œ ì €ì¥í•œë‹¤
    segment.export(output_filename, format="mp3")
    print(f"Success! Wrote audio to {output_filename}")

    # ì˜¤ë””ì˜¤ë¥¼ ì¬ìƒí•©ë‹ˆë‹¤.
    # Playë„ pydubì˜ íŒ¨í‚¤ì§€
    play(segment)
else:
    print(f"Failed to save file: {response.status_code}")
````

````python
# 5-5 ResNetì˜ ì‚¬ì „ í•™ìŠµëœ ëª¨ë¸ ì‚¬ìš©í•˜ê¸°

#python

# 1. ë¡œë“œ ë° ì¤€ë¹„

# fastaiì˜ ë¹„ì ¼ ë¼ì´ë¸ŒëŸ¬ë¦¬ì˜ ëª¨ë“ ê²ƒì„ ê°€ì ¸ì˜¨ë‹¤.
from fastai.vision.all import *

# ì¼ë°˜ì ìœ¼ë¡œ ë§ì´ ì‚¬ìš©í•˜ëŠ” íŠœí† ë¦¬ì–¼ ì½”ë“œ
# ë°ì´í„°ì…‹ ë¡œë“œ
# URLs.PETSëŠ” fastaiê°€ ì œê³µí•˜ëŠ” urlë¡œ ê³ ì–‘ì´ì™€ ê°•ì•„ì§€ ì´ë¯¸ì§€ë¥¼ í¬í•¨
path = untar_data(URLs.PETS)
# ì´ë¯¸ì§€ê°€ ì €ì¥ëœ images í´ë”ì˜ ê²½ë¡œ
# pathì— /imagesë¥¼ ì¶”ê°€í•´ ë””ë ‰í† ë¦¬ ë‚´ì˜ ì´ë¯¸ì§€ í´ë” ê²½ë¡œ ì§€ì •
path_imgs = path/'images'

# ì´ë¯¸ì§€ íŒŒì¼ ë¼ë²¨ë§ í•¨ìˆ˜ ì •ì˜
# ì´ í•¨ìˆ˜ëŠ” ì´ë¯¸ì§€ íŒŒì¼ì˜ ì´ë¦„ì„ ë°›ì•„ íŒë‹¨.
# ë¬¸ìì—´ xì˜ ì²« ë²ˆì§¸[0] ë¬¸ìë¥¼ ê°€ì ¸ì˜¨ë‹¤.
# .isupper()ëŠ” ëŒ€ë¬¸ìì¼ ë•Œ True ì†Œë¬¸ìì¼ ë•Œ Falseë¡œ ë°˜í™˜
def is_cat(x): return x[0].isupper()

# ë°ì´í„°ë¸”ë¡ ì •ì˜
# ImageDataLoaders.from_name_funcëŠ” ImageDataLoaders ê°ì²´ë¥¼ ìƒì„±í•˜ëŠ” ë©”ì„œë“œë¡œ, 
# ë°ì´í„°ë¥¼ í•™ìŠµìš©ê³¼ ê²€ì¦ìš©ìœ¼ë¡œ ë¶„í• í•˜ê³  ì „ì²˜ë¦¬
# path_imgs: ì´ë¯¸ì§€ ê²½ë¡œ
# get_image_files(path_imgs): ì´ë¯¸ì§€ íŒŒì¼ ë¦¬ìŠ¤íŠ¸ë¥¼ ê°€ì ¸ì˜¨ë‹¤.
# valid_pct=0.2: 20% ë°ì´í„°ë¥¼ ê²€ì¦ìš©ìœ¼ë¡œ ë¶„ë¦¬
# seed=42: ë¬´ì‘ìœ„ì„±ì„ ê³ ì •í•˜ì—¬ ë™ì¼í•œ ë¶„í•  ì„¤ì •
# label_func=is_cat: is_cat í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•´ ë¼ë²¨ì„ ìƒì„±
# item_tfms=Resize(224): ê° ì´ë¯¸ì§€ë¥¼ 224x224 í”½ì…€ë¡œ í¬ê¸° ì¡°ì •
dls = ImageDataLoaders.from_name_func(
    path_imgs, get_image_files(path_imgs), valid_pct=0.2, seed=42,
    label_func=is_cat, item_tfms=Resize(224))

# ë°ì´í„°ì…‹ í™•ì¸
# max_n=9ëŠ” í•œ ë²ˆì— 9ê°œì˜ ì´ë¯¸ì§€ ë³´ì´ê²Œ ì„¤ì •
# figsize=(7, 6)ëŠ” ì´ë¯¸ì§€ ì¶œë ¥ í¬ê¸°ë¥¼ ì§€ì •
dls.show_batch(max_n=9, figsize=(7, 6))



# 2. ì‚¬ì „í•™ìŠµëœ ResNet ëª¨ë¸ ë¡œë“œ ë° í•™ìŠµ
# cnn_learner: CNN ëª¨ë¸ì„ ì‚¬ìš©í•´ í•™ìŠµê¸°ë¥¼ ìƒì„±
# dlsëŠ” ë°ì´í„°ì…‹ ë¡œë” ì „ë‹¬
# resnet34 ì‚¬ì „ í•™ìŠµëœ ëª¨ë¸ì„ ì‚¬ìš©í•´ í•™ìŠµê¸° ìƒì„±
# metrics=error_rate: í•™ìŠµ ì¤‘ ì‚¬ìš©í•  ì„±ëŠ¥ í‰ê°€ ì§€í‘œë¥¼ ì„¤ì •(ì˜¤ë¥˜ìœ¨ ì‚¬ìš© ì§€ì •)
learn = cnn_learner(dls, resnet34, metrics=error_rate)

# í•™ìŠµë¥  ì°¾ê¸° (ìµœì ì˜ í•™ìŠµë¥ ì„ ìë™ìœ¼ë¡œ ì°¾ì•„ì¤Œ)
# learn.lr_find(): ìµœì ì˜ í•™ìŠµë¥ ì„ ì°¾ëŠ” í•¨ìˆ˜
learn.lr_find()

# ëª¨ë¸ í•™ìŠµ (ì‚¬ì „ í•™ìŠµëœ ëª¨ë¸ì— ëŒ€í•´ íŒŒì¸ íŠœë‹)
# learn.fine_tune(3): ì‚¬ì „ í•™ìŠµëœ ëª¨ë¸ì„ ê¸°ë°˜ìœ¼ë¡œ 3 epoch ë™ì•ˆ íŒŒì¸ íŠœë‹ì„ ì§„í–‰
learn.fine_tune(3)




# 3. ëª¨ë¸ í‰ê°€
# show_results: ì˜ˆì¸¡ê²°ê³¼ ì‹œê°í™”
learn.show_results()

# í˜¼ë™ í–‰ë ¬ (Confusion Matrix) ì¶œë ¥
# ClassificationInterpretation.from_learner: í•™ìŠµëœ ëª¨ë¸ì„ í•´ì„í•˜ê³ , í˜¼ë™ í–‰ë ¬ì„ ìƒì„±
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
````
