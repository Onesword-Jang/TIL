# 문자열에 사용 가능한 함수

## 대소문자 변환
- **capitalize()**
 - 첫 번째 문자를 대문자로 변환합니다.
- **casefold()**
 - 문자열을 소문자로 변환합니다
- **lower()**
 - 문자열을 소문자로 변환합니다.
- **upper()**
 - 문자열을 대문자로 변환합니다.
- **swapcase()**
 - 대소문자를 바꾸어 소문자를 대문자로, 그 반대의 경우도 마찬가지입니다.
- **title()**
 - 각 단어의 첫 문자를 대문자로 변환합니다.


## 문자열 반환(출력)하기
- **center()**
 - 중앙에 위치한 문자열을 반환합니다.
- **count()**
 - 문자열에서 지정된 값이 발생한 횟수를 반환합니다.
- **encode()**
 - 문자열의 인코딩된 버전을 반환합니다.
- **endswith()**
 - 문자열이 지정된 값으로 끝나면 참을 반환합니다.
- **find()**
 - 문자열에서 지정된 값을 검색하고 찾은 위치의 위치를 반환합니다.
- **index()**
 - 문자열에서 지정된 값을 검색하고 해당 값을 찾은 위치를 반환합니다.
- **lstrip()**
 - 문자열의 왼쪽 자르기 버전을 반환합니다.
- **maketrans()**
 - 번역에 사용할 번역 테이블을 반환합니다.
- **partition()**
 - 문자열이 세 부분으로 분할된 튜플을 반환합니다.
- **replace()**
 - 지정된 값이 지정된 값으로 대체된 문자열을 반환합니다.
- **join()**
 - 이터러블의 요소를 문자열로 변환합니다.
- **rfind()**
 - 문자열에서 지정된 값을 검색하고 해당 값이 발견된 마지막 위치를 반환합니다.
- **rindex()**
 - 문자열에서 지정된 값을 검색하고 해당 값이 발견된 위치의 마지막 위치를 반환합니다.
- **rjust()**
 - 문자열의 오른쪽 정렬 버전을 반환합니다.
- **rpartition()**
 - 문자열이 세 부분으로 분할된 튜플을 반환합니다.
- **rsplit()**
 - 지정된 구분 기호에서 문자열을 분할하고 목록을 반환합니다.
- **rstrip()**
 - 문자열의 오른쪽 트림 버전을 반환합니다.
- **split()**
 - 지정된 구분 기호에서 문자열을 분할하고 목록을 반환합니다.
- **splitlines()**
 - 줄 바꿈에서 문자열을 분할하고 목록을 반환합니다.
- **startswith()**
 - 문자열이 지정된 값으로 시작하면 참을 반환합니다.
- **strip()**
 - 문자열의 잘린 버전을 반환합니다.
- **translate()**
 - 번역된 문자열을 반환합니다.
- **zfill()**
 - 문자열의 처음에 지정된 수의 0 값으로 문자열을 채웁니다.


## 출력값 bool으로 반환
- **isalnum()**
 - 문자열의 모든 문자가 영숫자인 경우 True를 반환합니다.
- **isalpha()**
 - 문자열의 모든 문자가 알파벳이면 True를 리턴합니다.
- **isascii()**
 - 문자열의 모든 문자가 아스키 문자이면 True를 리턴합니다.
- **isdecimal()**
 - 문자열의 모든 문자가 10진수이면 True를 리턴합니다.
- **isdigit()**
 - 문자열의 모든 문자가 숫자인 경우 True를 리턴합니다.
- **isidentifier()**
 - 문자열이 식별자인 경우 True를 리턴합니다.
- **islower()**
 - 문자열의 모든 문자가 소문자인 경우 True를 리턴합니다.
- **isnumeric()**
 - 문자열의 모든 문자가 숫자인 경우 참을 리턴합니다.
- **isprintable()**
 - 문자열의 모든 문자가 인쇄 가능하면 True를 리턴합니다.
- **isspace()**
 - 문자열의 모든 문자가 공백이면 참을 리턴합니다.
- **istitle()**
 - 문자열이 제목의 규칙을 따르는 경우 True를 리턴합니다.
- **isupper()**
 - 문자열의 모든 문자가 대문자인 경우 True를 반환합니다.
- **ljust()**
 - 문자열의 왼쪽 정렬 버전을 반환합니다.


## 설정 및 지정하기
- **expandtabs()**
 - 문자열의 탭 크기를 설정합니다.
- **format()**
 - 문자열에서 지정된 값의 서식을 지정합니다.
- **format_map()**
 - 문자열에서 지정된 값의 형식을 지정합니다.
